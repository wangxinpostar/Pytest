```python
# 1. 读取数据，对数据进行统计分析与可视化
#（1）读取salary.csv中的数据集，用散点图绘制工龄与薪资的分布图，并打印读取的前5行数据。

import pandas as pd
import matplotlib.pyplot as plt

# 读取csv文件
data = pd.read_csv('E5-salary.csv')

# 取出工龄和薪资数据
x = data['Year']
y = data['Salary']

# 绘制散点图
plt.xlabel(u'Year')
plt.ylabel(u'Salary')
plt.scatter(x, y)

# 打印前5行数据
print('前五行的数据为')
data.head(5)
```

```python
#（2）去除重复和有缺失项的数据行，用散点图绘制工龄与薪资的分布图，并打印结果的前5行。
#去除缺少项数据行
data.dropna()
#去除重复项数据行
data.drop_duplicates()
x=data['Year']
y=data['Salary']
plt.xlabel(u'Year')
plt.ylabel(u'Salary')
plt.scatter(x,y)
print('前五行的数据为')
data.head(5)
```

```python
# （3）筛选工龄小于12年，工资小于12万的数据行，用散点图绘制工龄与薪资的分布图，并打印结果的前5行。
data=data[data['Year']<12]
data=data[data['Salary']<120000]
data.head(5)
```

```python
#（4）统计每个工龄的平均工资，用柱状图展现每个工龄的平均工资，并打印结果的前5行。
# 以int类型的变量来统计工龄
group_data=data.groupby(data['Year'].apply(lambda x:int(x)))['Salary'].mean()
group_data.head(5)
```

```python
# （5）用柱状图、散点图分别展现每个工龄的平均工资，要求使用用上下子图形式展现
# 计算每个工龄的平均工资
group_data = data.groupby('Year')['Salary'].mean()

# 将工龄和平均工资分别赋值给 x 和 y
x=group_data.index
y=group_data.values

# 使用 subplot 将两个图形展现在上下两个子图中
plt.subplot(2,1,1)
plt.bar(x,y)  # 画柱状图
plt.title('Average Salary by Year')  # 添加标题
plt.xlabel('Year')  # 添加 x 轴标签
plt.ylabel('Salary')  # 添加 y 轴标签

plt.subplot(2,1,2)
plt.scatter(x,y)  # 画散点图
plt.title('Average Salary by Year')  # 添加标题
plt.xlabel('Year')  # 添加 x 轴标签
plt.ylabel('Salary')  # 添加 y 轴标签

plt.tight_layout()  # 调整子图之间的间距
plt.show()
```

```python
from sklearn import linear_model
import numpy as np

# 将Year和Salary列分别转换为NumPy数组，用于sklearn中的线性回归模型
x = data['Year'].to_numpy()[:, np.newaxis]
y = data['Salary'].to_numpy()[:, np.newaxis]

# 创建线性回归模型对象
regr = linear_model.LinearRegression()

# 将样本数据拟合到线性模型中
regr.fit(x, y)

# 计算模型的拟合度分数
regr_score = regr.score(x, y)
print('线性回归预测的结果:', regr_score)

# 绘制散点图，展示样本数据和线性回归模型的预测结果
plt.subplots()
plt.scatter(x, regr.predict(x), color='red', label=u'predict')
plt.scatter(x, y, color='blue', label=u'data')
plt.legend()
plt.show()
```

```python
#（2）采用sklearn包中的三阶多项式拟合模型对工龄与薪资之间的关系进行回归预测
# 绘制预测结果时用散点图表示样本数据的分布，并绘制三阶多项式预测的结果。
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

x=data['Year'].to_numpy()[:,np.newaxis]
y=data['Salary'].to_numpy()[:,np.newaxis]

poly_reg = PolynomialFeatures(degree=3)
x_data_poly = poly_reg.fit_transform(x)
model = LinearRegression()
model.fit(x_data_poly, y)
model_score=model.score(x_data_poly,y)
print('三阶多项式预测的结果:',model_score)

plt.scatter(x,y,color='blue',label=u'data')
plt.scatter(x,model.predict(x_data_poly),color='red',label='predict')
plt.legend()
plt.show()
```

```python
# 导入需要的库和模块
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 读取数据
data = pd.read_csv('salary.csv')

# 获取特征和目标变量
x = data['Year'].to_numpy()[:, np.newaxis]
y = data['Salary'].to_numpy()[:, np.newaxis]

# 将数据集切分为训练集（70%）和测试集（30%）
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3)

# 线性回归预测
regr = LinearRegression()
regr.fit(x_train, y_train)
regr_score = regr.score(x_train, y_train)
print("训练集的线性回归模型评估指标：", regr_score)
y_pred = regr.predict(x_test)

# 绘制测试集数据工龄与薪资折线图
plt.scatter(x_test, y_test, color='blue', label='test')
plt.plot(x_test, y_pred, color='red', label='pred')
plt.title('Test Data: Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# 三阶多项式回归预测
poly_reg = PolynomialFeatures(degree=3)
x_data_poly = poly_reg.fit_transform(x_train)
model = LinearRegression()
model.fit(x_data_poly, y_train)
model_score = model.score(x_data_poly, y_train)
print("训练集的三阶多项式回归模型评估指标：", model_score)
x_test_poly = poly_reg.fit_transform(x_test)

# 绘制线性回归和三阶多项式回归在测试数据集上预测的结果中工龄与薪资的折线图
plt.scatter(x_test, y_test, color='blue', label='data')
plt.plot(x_test, regr.predict(x_test), color='green', label='linear')
plt.plot(x_test, model.predict(x_test_poly), color='red', label='poly')
plt.title('Test Data: Years of Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

**题目2（20分）：** 利用KNN分类算法对“威斯康兴乳腺癌数据分析及自动诊断”数据进行分类处理，该数据存储在**“E5-bc_data.csv”**中，数据中“diagnosis”为诊断结果，即分类标签，从“radius_mean”列到最后一列为检查数据，即用于分类的依据。 

```python
# （1）读取数据，并显示前10行的数据
import pandas as pd
bc_data=pd.read_csv('E5-bc_data.csv')
bc_data.head(10)
```

```python
# （2）数据理解，查看数据的形状和描述性统计
bc_data.describe()
```

```python
#（3）数据准备
group_data=bc_data.groupby(by='diagnosis')
group_data.head()
```

```python
#（4）数据建模和训练，要求对样本数据进行切分，分为训练集和测试集
X = bc_data.drop('diagnosis', axis=1)  # 特征数据
y = bc_data['diagnosis']  # 目标变量数据

from sklearn.model_selection import train_test_split

# 将数据集分为训练集和测试集，其中测试集占比为25%，随机数种子为6
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=6)
```

```python
#（5）用训练后的模型对测试集进行预测
from sklearn.neighbors import KNeighborsClassifier

# 定义KNN分类器
model = KNeighborsClassifier()

# 使用训练集进行模型训练
model.fit(X_train, y_train)

from sklearn.metrics import classification_report

# 输出训练集的模型评估指标
print("训练集的模型评估指标：")

# 输出模型在训练集上的准确率
model_score = model.score(X_train, y_train)
print('The accuracy of train data is', model_score)
print('--------------------------------------------------------------------------')

# 对训练集进行预测并输出分类报告
y_train_predict = model.predict(X_train)
model_report1 = classification_report(y_train, y_train_predict)
print(model_report1)
print('$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$')

# 输出测试集的模型评估指标
print("测试集的模型评估指标：")

# 输出模型在测试集上的准确率
model_score = model.score(X_test, y_test)
print('The accuracy of test data is', model_score)
print('--------------------------------------------------------------------------')

# 对测试集进行预测并输出分类报告
y_predict = model.predict(X_test)
model_report = classification_report(y_test, y_predict)
print(model_report)
print('--------------------------------------------------------------------------')
```

**题目3（20分）：** 利用K-means聚类算法对“不同国家蛋白质消费结构”数据进行分析。根据蛋白质摄入情况，分析各个国家的饮食习惯的相似性，该数据存储在**“E5-protein.txt”**中。

```python
# 读取数据，并显示前10行的数据
protein =  pd.read_table('E5-protein.txt',sep='\t')
protein.head()
```

```python
# 对数据进行归一化处理
from sklearn.preprocessing import minmax_scale
from sklearn.cluster import KMeans  
protein_data=minmax_scale(protein.drop('Country',axis=1))
```

```python
# 数据建模和训练，采用K-means聚类算法进行建模聚类
# 注意K的取值，可以多测试几个K值做比较
estimator = KMeans(n_clusters = 3) 
#构造K-Means聚类模型
estimator.fit(protein_data)   
#数据导入模型进行训练
label_pred = estimator.labels_   
```

```python
# 输出聚类结果
print(label_pred)
```

**题目4（20分）：** 参考讲义中的人脸检测和识别代码，自选一张包含自己人脸的图像，对图像中的人脸、眼睛、鼻子和笑脸进行检测。  
注意：此题回答时将识别结果的图像插入到该文档，不提交原始图像。

```python
import numpy as np
import cv2 as cv

# 加载人脸分类器
face_cascade = cv.CascadeClassifier('./shibie/cascade_files/haarcascade_frontalface_alt.xml')
# 加载微笑分类器
smile_cascade = cv.CascadeClassifier('./shibie/cascade_files/haarcascade_smile.xml')
# 加载眼睛分类器
eye_cascade = cv.CascadeClassifier('./shibie/cascade_files/haarcascade_eye.xml')

# 读取图片
img = cv.imread('people.jpg')
# 将图片转为灰度图
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# 检测人脸
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    # 绘制矩形框
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    # 获取人脸区域的灰度图和彩色图
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    # 检测鼻子
    noses = nose_cascade.detectMultiScale(roi_gray,1.3,5)
    for (ex,ey,ew,eh) in noses:
        # 绘制鼻子矩形框
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        break
    # 检测眼睛
    eyes= eye_cascade.detectMultiScale(roi_gray,1.6,5)
    for (ex,ey,ew,eh) in eyes:
        # 绘制眼睛矩形框
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
    # 检测微笑
    smile = smile_cascade.detectMultiScale(roi_gray,1.5,5)
    for (ex,ey,ew,eh) in smile:
        # 绘制微笑矩形框
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)
        break

# 显示结果图片
cv.imshow('Eyes and Smile Detector...',img)
# 等待按键
cv.waitKey(0)
# 销毁所有窗口
cv.destroyAllWindows()
```
