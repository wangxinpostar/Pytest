**答：**  
1.	Matplotlib：Matplotlib是Python中最常用的数据可视化库之一，它提供了多种绘图方式，包括线图、柱状图、散点图、饼图等。

各种技术数据可视化绘图的基本方法：
1.	Matplotlib绘图方法：使用pyplot模块中的函数进行绘图，常用函数包括plot()、scatter()、bar()、pie()等。
2.	Seaborn绘图方法：使用seaborn库中的函数进行绘图，常用函数包括heatmap()、catplot()、lineplot()、scatterplot()等。
3.	Plotly绘图方法：使用plotly库中的函数进行绘图，常用函数包括scatter()、line()、bar()、pie()等。
4.	Bokeh绘图方法：使用bokeh库中的函数进行绘图，常用函数包括scatter()、line()、bar()、pie()等。

主要的可视化图形种类包括：
1.	线图：展示数据的变化趋势和趋势变化的速度。
2.	柱状图：展示各个类别或数据点之间的数量或大小的差异。
3.	散点图：展示两个变量之间的相关性或分布情况。
4.	饼图：展示各个部分的占比或比例。
5.	热图：展示数据在二维平面上的分布情况和密度。
6.	3D图：展示数据在三维空间中的分布和变化情况。
7.	树状图：展示数据之间的层次结构和关系。

**题目2，基于Matplotlib模块的数据可视化（35分）：**  
针对一组关于全球星巴克门店的统计书，分析在不同国家和地区以及中国不同城市的星巴克门店的数量。分析步骤如下：
1. 导入模块  
2. 数据读取
3. 数据分析与可视化：  
（1）查看星巴克旗下有哪些品牌。如果只关心星巴克咖啡门店，则只需要获取星巴克种Brand的数据集，并查看全世界一共有多少家星巴克门店；  
（2）查看全世界一共有多少个国家和地区开设了星巴克门店，显示门店数量排名前10和后10的国家和地区；  
（3）用柱状图绘制排名前10的分布情况；  
（4）按照星巴克门店在中国的分布情况，统计显示排名前10的城市，并用柱状图绘制在中国排名前10的分布情况；  
（5）用饼状图显示星巴克门店的经营方式有哪几种。

```python
# 题目2
# 1. 导入相应的模块
from pylab import mpl
import pandas as pd
import numpy as np
from matplotlib.font_manager import FontProperties 
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.sans-serif'] = ['KaiTi']
mpl.rcParams["axes.unicode_minus"] = False
%matplotlib inline
```

```python
# 2. 数据读取
df=pd.read_csv(r"directory.csv")
display(df)
```

```python
# 3. 数据分析与可视化
#（1）查看星巴克旗下有哪些品牌。如果只关心星巴克咖啡门店，则只需要获取星巴克种Brand的数据集，并查看全世界一共有多少家星巴克门店
starbucks = df[df['Brand'] == 'Starbucks']
total_count = len(starbucks)
print('全球星巴克门店数量：', total_count)
```

```python
#（2）查看全世界一共有多少个国家和地区开设了星巴克门店，显示门店数量排名前10和后10的国家和地区
count_by_country = starbucks.groupby('Country')['Brand'].count().reset_index()
count_by_country.sort_values(by='Brand', ascending=False, inplace=True)
top10 = count_by_country.head(10)
bottom10 = count_by_country.tail(10)
print('全球星巴克门店所在国家和地区数量：', len(count_by_country))
print('排名前10的国家和地区：')
display(top10)
print('排名后10的国家和地区：')
display(bottom10)
```

```python
#（3）用柱状图绘制排名前10的分布情况
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
plt.title('Top 10 Countries/Regions with Most Starbucks Stores')
sns.barplot(x='Country', y='Brand', data=top10, palette='Blues_d')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country/Region')
plt.ylabel('Number of Starbucks stores')
```

```python
#（4）按照星巴克门店在中国的分布情况，统计显示排名前10的城市，并用柱状图绘制在中国排名前10的分布情况
# 筛选出Brand为Starbucks、Country为China的数据
starbucks_in_china = df[(df['Brand'] == 'Starbucks') & (df['Country'] == 'US')]
# 统计在中国各城市的星巴克门店数量及排名前10的城市
count_by_city = starbucks_in_china.groupby('City')['Brand'].count().reset_index()
top10 = count_by_city.sort_values(by='Brand', ascending=False).head(10)
# 绘制柱状图
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
plt.title('Top 10 Cities with Most Starbucks Stores in China')
sns.barplot(x='City', y='Brand', data=top10, palette='Blues_d')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.xlabel('City', fontsize=14)
plt.ylabel('Number of Starbucks stores', fontsize=14)
```

```python
#（5）用饼状图显示星巴克门店的经营方式有哪几种
count_by_ownership = starbucks.groupby('Ownership Type')['Brand'].count().reset_index()
# 绘制饼状图
plt.figure(figsize=(8, 8))
plt.title('Starbucks Store Ownership Type')
plt.pie(count_by_ownership['Brand'], labels=count_by_ownership['Ownership Type'], autopct='%1.1f%%')
plt.show()
```
