```python
import requests
from requests.structures import CaseInsensitiveDict

from IPython.display import display
from IPython.core.display import HTML

book_info = []
page = 1
while True:
    next_page_url = f"https://book.douban.com/tag/%E8%AE%A1%E7%AE%97%E6%9C%BA?start={(page-1)*20}&type=T"

    headers = CaseInsensitiveDict()
    headers["User-Agent"] = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.80 Safari/537.36 Edg/98.0.1108.50"

    resp = requests.get(next_page_url, headers=headers)

    from lxml import etree
    s = etree.HTML(resp.text)

    for block in s.xpath("//li[@class='subject-item']"):
        title = ''
        elem_title = block.xpath('.//h2/a')
        if elem_title:
            title = ''.join(elem_title[0].itertext()).replace(' ', '').replace('\n', '')

        price = -1.0
        elem_price = block.xpath(".//span[@class='buy-info']/a/text()")
        if elem_price:
            tmp = elem_price[0].strip()
            price = float(tmp[tmp.find('版')+2 : tmp.find('元')])

        coverurl = ''
        elem_coverurl = block.xpath('.//img/@src')
        if elem_coverurl:
            cover = elem_coverurl[0]

        print(title, price)

        book_info.append([title, price, cover])

    print(next_page_url)

    page = page + 1
    if page > 4:
        break
```

```python
from lxml import etree
import requests
from requests.structures import CaseInsensitiveDict

from IPython.display import display
from IPython.core.display import HTML

page = 1

while True:
    next_page_url = f"https://api.bilibili.com/x/web-interface/popular?ps=20&pn={page}"

    resp = requests.get(next_page_url)
    content_json_format = resp.json()
    for item in content_json_format['data']['list']:
        print(item['tname'])

    page = page + 1
    if page > 3:
        break
```
