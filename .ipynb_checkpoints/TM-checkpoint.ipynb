```python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
cid = 1052723919

def getData(cid: int):
    url = 'https://comment.bilibili.com/{}.xml'.format(cid)
    rq = requests.get(url)
    rq.encoding = 'utf-8'
    soup = BeautifulSoup(rq.text, 'lxml')
    tmp = soup.select('d')
    comments = [i.text for i in tmp]
    info_else = [i.get('p').split(',') for i in tmp]
    columns = ['出现时间点', ' 模式 ', ' 字体 ', ' 颜色 ', '发送时间 ', ' 弹幕池 ', ' 用户 ID', 'rowID','number']
    data = pd.DataFrame(info_else, columns=columns)
    data['comment'] = comments
    data.to_csv('data.csv', index=None, encoding='utf-8-sig')
    print("已经完成了!!!!!!!")
    return 
getData(cid)
```

```python
import jieba 

def getStopWords():
    """
    得到停用词
    """
    with open("cn_stopwords.txt", encoding="utf-8") as file:
        stop_words = file.read()
        stop_words = stop_words.split("\n")
    return stop_words

def getString():
    data = pd.read_csv("data.csv")["comment"]
    return "".join(data)

def shanchu(data, stop_words):
    res = []
    for x in data:
        if x not in stop_words and x != " " and len(x) > 1:
            res.append(x)
    return res
    

def count(df):
    dt = {}
    for i in df:
        if i in dt:
            dt[i] += 1
        else:
            dt[i] = 1
    dt=sorted(dt.items(), key=lambda x :x[1], reverse=True)
    return dt


data = getString()
data_list = jieba.cut(data)
st = getStopWords()
df = shanchu(data_list, st)
df = count(df)
```

```python
# 转换为字典类型
word_dict = dict(df)

# 生成词云图
wc = WordCloud(
    background_color="white",
    max_words=2000,
    width=1024,
    height=768,
    font_path='C:\Windows\Fonts\STFANGSO.TTF'
)
wc.generate_from_frequencies(word_dict)

# 显示词云图
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show() 
```

```python
data_sorted = sorted(df, key=lambda x: x[1], reverse=True)

# Get the top 50 words and their frequencies
top_words = [x[0] for x in data_sorted[:25]]
top_freqs = [x[1] for x in data_sorted[:25]]
display(list(zip(top_freqs,top_words)))

# Plot the bar chart
plt.bar(top_words, top_freqs)
plt.xticks(rotation=90)
plt.xlabel('Words')
plt.ylabel('Frequencies')
plt.title('Top 25 Chinese Words')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读入数据
df = pd.read_csv("data.csv")
user_counts = df[" 用户 ID"].value_counts()

# 统计每个发送弹幕次数的用户数量
counts = {}
for i in range(1, 8):
    counts[i] = (user_counts == i).sum()
counts[7] = (user_counts >= 7).sum()

display(list(zip(counts.keys(), counts.values())))
print(sum(counts.values()))
# 绘制柱状图
plt.bar(counts.keys(), counts.values())
plt.bar(counts.keys(), counts.values(), label="弹幕数")
plt.xlabel("发送弹幕次数")
plt.ylabel("用户数量")
plt.title("不同发送弹幕次数的用户数量")
plt.legend()
plt.show()
```

```python
df = pd.read_csv("data.csv")

# 将时间点转换为整数
df["出现时间点"] = df["出现时间点"].astype(float).astype(int)

# 按照时间排序
df = df.sort_values(by=["出现时间点"])

# 设置时间段的长度（单位：秒）
interval = 60

# 统计每个时间段内的弹幕数量
time_points = range(0, df["出现时间点"].max()+interval, interval)
num_comments = [((df["出现时间点"]>=start) & (df["出现时间点"]<start+interval)).sum() for start in time_points]

# 绘图
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(time_points[:-1], num_comments[:-1], width=interval*0.8, align="edge")
ax.set_xlabel("时间（秒）")
ax.set_ylabel("弹幕数量")
ax.set_title("弹幕发布数量在影片时间刻度上的分布柱状图")
plt.show()
```
