**题目1（10分）：**  
简述numpy模块中的数组、Pandas模块中的Series和DataFrame数据结构各自的特点，以及他们与Python自带的列表、元组、字典等数据类型的区别。

**答：**  
(一)	NumPy数组和Python列表的区别：
1.	NumPy数组可以容纳同一数据类型的元素，而Python列表可以容纳不同类型的元素。
2.	NumPy数组支持向量化操作，使得对数组的操作可以在整个数组上进行，而不需要使用显式循环，从而提高了计算效率。
3.	NumPy数组支持广播（Broadcasting）功能，可以对形状不同的数组进行算术运算，使得编写代码更加简洁。
4.	NumPy数组占用的内存比Python列表更少，因为NumPy数组中的元素占用的空间更小。
(二)	Pandas中的Series和DataFrame与NumPy数组的区别：
1.	Series是一维的数据结构，类似于Python的字典，其中包含索引和数据，而DataFrame是二维的数据结构，类似于电子表格，由多个Series组成。
2.	Series和DataFrame可以容纳不同类型的元素，但是由于Pandas是基于NumPy开发的，因此它们在性能和效率方面具有NumPy的特点。
3.	Series和DataFrame具有许多内置的方法和功能，使得数据分析和处理变得更加简单和高效。
4.	Series和DataFrame可以对缺失的数据进行处理，提供了灵活的处理方式。
(三)	与Python自带的列表、元组、字典等数据类型的区别：
1.	列表可以容纳不同类型的元素，而NumPy数组和Pandas的Series和DataFrame只能容纳同一类型的元素。
2.	列表、元组和字典不支持向量化操作，而NumPy数组和Pandas的Series和DataFrame支持向量化操作。
3.	列表和元组是不可变的，而NumPy数组和Pandas的Series和DataFrame是可变的。
4.	字典是由键和值组成的，而Pandas的Series是由索引和值组成的。DataFrame是由多个Series组成，可以看作是由多个字典组成的。

**题目2（25分）：**  
编写程序，采用numpy模块中的随机数生成函数（随机数的种子数设置为10）创建一个数字大小范围为0~100，形状为（10，10）的二维数组 myarr1，然后对数组myarr1进行以下操作：  
（1）获取数组myarr1前5行的数据并输出；  
（2）从数组myarr1的第1行开始到第7行，步长为2进行切片，并输出切片后的结果；  
（3）获取数组myarr1第3行和第7行中，第2到第6列的数据并输出；  
（4）将数组myarr1纵向拆分成两部分，上半部分为数组myarr1的前4行，下半部分为数组myarr1的最后6行；  
（5）将数组myarr1转换为形状为（5，20）的数组，并输出转换后的结果。

```python
# 题目2
import numpy as np

# 设置随机数种子数为 10，生成形状为 (10, 10) 的二维数组 myarr1
np.random.seed(10)
myarr1 = np.random.randint(0, 100, (10, 10))

# 获取数组 myarr1 的前 5 行的数据并输出
print("前 5 行数据：")
display(myarr1[:5])

# 对数组 myarr1 进行切片，从第 1 行开始到第 7 行，步长为 2，并输出切片后的结果
print("切片后的结果：")
display(myarr1[0:8:2])

# 获取数组 myarr1 第 3 行和第 7 行中，第 2 到第 6 列的数据并输出
print("第 3 行和第 7 行，第 2 到第 6 列的数据：")
display(myarr1[[2, 7]][:, 1:6])

# 将数组 myarr1 纵向拆分成两部分，上半部分为数组 myarr1 的前 4 行，下半部分为数组 myarr1 的最后 6 行
myarr1_split = np.vsplit(myarr1, [4])

# 输出拆分后的结果
print("拆分后的结果：")
display(myarr1_split)

# 将数组 myarr1 转换为形状为 (5, 20) 的数组，并输出转换后的结果
myarr1_reshaped = np.reshape(myarr1, (5, 20))
print("转换后的结果：")
display(myarr1_reshaped)
```

**题目3（25分）：**  
假设有一组鸢尾花花瓣长度的数据：1.0, 2.1, 1.2, 3.3, 2.4, 1.5, 4.6, 5.7, 1.9, 3.0, 3.3, 4.5, 3.7, 3.4, 3.8, 3.9, 4.1, 4.2, 6.0, 3.1, 6.3。  
编写程序，采用numpy模块中的相关函数对这组数据进行排序，并求出和、累积和、均值、标准差、方差、最小值、最大值。  
要求输出这些结果，输出时要有相应的提示性语句（例如：print("这组鸢尾花花瓣长度数据的均值为：", meanvalue)）。

```python
# 题目3
import numpy as np

# 定义鸢尾花花瓣长度数据
data = np.array([1.0, 2.1, 1.2, 3.3, 2.4, 1.5, 4.6, 5.7, 1.9, 3.0, 3.3, 4.5, 3.7, 3.4, 3.8, 3.9, 4.1, 4.2, 6.0, 3.1, 6.3])

# 对数据进行排序
sorted_data = np.sort(data)

# 求和
sum_value = np.sum(sorted_data)

# 累积和
cumulative_sum = np.cumsum(sorted_data)

# 均值
mean_value = np.mean(sorted_data)

# 标准差
std_value = np.std(sorted_data)

# 方差
var_value = np.var(sorted_data)

# 最小值
min_value = np.min(sorted_data)

# 最大值
max_value = np.max(sorted_data)

# 输出结果
print("排序后的数据为：", sorted_data)
print("这组鸢尾花花瓣长度数据的和为：", sum_value)
print("这组鸢尾花花瓣长度数据的累积和为：", cumulative_sum)
print("这组鸢尾花花瓣长度数据的均值为：", mean_value)
print("这组鸢尾花花瓣长度数据的标准差为：", std_value)
print("这组鸢尾花花瓣长度数据的方差为：", var_value)
print("这组鸢尾花花瓣长度数据的最小值为：", min_value)
print("这组鸢尾花花瓣长度数据的最大值为：", max_value)
```

**题目4（40分）：**  
使用Pandas模块加载测试数据“testDataE1.xlsx”，然后对该数据完成以下操作：  
（1）加载数据“testDataE1.xlsx”，并显示该数据的前10行；  
（2）用方法describe()查看该数据框的描述性统计；  
（3）查看该数据框的形状；  
（4）采用显式index读取“白细胞计数”这一列的前5行；  
（5）提取第30至45行中“淋巴细胞计数”和“血小板计数”这两列的数据；  
（6）提取第2至6列的所有行数据；  
（7）对该数据按“血小板计数”这一列进行降序排序；  
（8）利用groupby()函数进行分组统计：以“性别”作为分组依据或条件；分组后统计“白细胞计数”的平均值、最大值、中值、最小值。

```python
# 题目4
import pandas as pd
pd.set_option('display.max_rows', None)  # 显示所有行
pd.set_option('display.max_columns', None)  # 显示所有列
# (1) 加载数据并显示前10行
df = pd.read_excel('testDataE1.xlsx')
display(df.head(10))

# (2) 用方法describe()查看该数据框的描述性统计
display(df.describe())

# (3) 查看该数据框的形状
display(df.shape)

# (4) 采用显式index读取“白细胞计数”这一列的前5行
display(df.loc[0:4, '白细胞计数'])

# (5) 提取第30至45行中“淋巴细胞计数”和“血小板计数”这两列的数据
display(df.loc[29:45, ['淋巴细胞计数', '血小板计数']])

# (6) 提取第2至6列的所有行数据
display(df.iloc[:, 1:6])

# (7) 对该数据按“血小板计数”这一列进行降序排序
display(df.sort_values(by='血小板计数', ascending=False))

# (8) 利用groupby()函数进行分组统计
grouped = df.groupby('性别')['白细胞计数']
display(grouped.agg(['mean', 'max', 'median', 'min']))
```
