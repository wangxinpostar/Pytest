```python
import pandas as pd
Tv=pd.read_excel('DaPy_data.xlsx','Target',index_col=0)
Tv #目标值 
```

```python
Tv['年收益']=Tv.年销售量*(Tv.销售单价-Tv.单件成本)-Tv.设备投资
Tv
```

```python
import matplotlib.pyplot as plt 
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['figure.dpi']=90 #分辨率
Tv['年收益'].plot(kind='bar')
```

```python
Tv['年收益'].idxmax()
```

```python
Ev=[min(Tv.设备投资), min(Tv.单件成本), max(Tv.年销售量), max(Tv.销售单价), 
 max(Tv.年收益)]
Ev #理想值 
```

```python
Tv_Ev2=((Tv-Ev))**2 #差值的平方 
Tv_Ev2
```

```python
Dv=(Tv_Ev2).sum(1)
Dv #差距
```

```python
Tv['差距']=Dv
Tv
```

```python
Dv.plot(kind='bar')
```

```python
Dv.idxmin() 
```

```python
pd.DataFrame({'单目标':Tv['年收益'],'单目标方案':Tv['年收益']==Tv['年收益'].max(),
 '多目标':Tv['差距'],'多目标方案':Tv['差距']==Tv['差距'].min()})
```

```python
# 根据条件，可得到损益矩阵
PLm=pd.DataFrame() #构建损益矩阵 ProfitLoss matrix 
PLm['畅销']= 12000*(Tv.销售单价-Tv.单件成本)-Tv.设备投资
PLm['一般']= 8000*(Tv.销售单价-Tv.单件成本)-Tv.设备投资
PLm['滞销']= 1500*(Tv.销售单价-Tv.单件成本)-Tv.设备投资
PLm
```

```python
lg=PLm.max(axis=1)
lg #每列最大者
```

```python
from IPython.core.interactiveshell import InteractiveShell as IS
IS.ast_node_interactivity = "all"
```

```python
BQD=PLm.copy()
PLm
BQD['乐观']=lg
BQD
```

```python
lg.plot(kind='bar')
```

```python
bg=PLm.min(1)
bg
```

```python
BQD['悲观']=bg
BQD
```

```python
plt.rcParams['axes.unicode_minus']=False #正常显示图中负号
bg.plot(kind='bar')
```

```python
bg.idxmax() 
```

```python
a=0.35
zz= a*lg + (1-a)*bg
zz
```

```python
BQD['折中']=zz
BQD
```

```python
zz.plot(kind='bar')
```

```python
zz.idxmax()
```

```python
Rm=PLm.max()-PLm
Rm #构建后悔矩阵 Regret matrix 
```

```python
hh=Rm.max(1)
hh
```

```python
BQD['后悔']=hh
BQD
```

```python
hh.plot(kind='bar')
```
